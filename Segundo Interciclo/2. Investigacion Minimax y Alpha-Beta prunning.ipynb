### Algoritmo minimax

El algoritmo de bsqueda Minimax, es una bsqueda de profundidad limitada. El nombre del algoritmo se da al considerar que dada una funcin esttica que devuelve valores en relacin al jugador maximizante, este trata de maximizar su valor mientras que el oponente trata de minimizarlo. En un rbol de juego donde los valores de la funcin esttica, estn en relacin con el jugador maximizante, se maximiza y minimiza de forma alterna de un nivel a otro.

Entre las principales caractersticas que posee este algoritmo tenemos:

- Facilidad de problemas complejos con reglas simples.
- Pruebas contra humanos escalables
- Existencia de un solo ganador
- Exploracin de N capas
- Tiempo de exploracin agotado
- Situaciones estticas sin cambios significativos

### Algoritmo Poda Alfa-Beta

El algoritmo poda Alfa-Beta como ya lo habamos mencionado anteriormente es una tcnica mejorada del algoritmo Minimax en la cual es posible calcular un estado objetivo sin la necesidad de recorrer todos los nodos del rbol de juegos, este tipo de algoritmo suele utilizarse para cualquier tipo de rbol de bsqueda en profundidad y para subrboles enteros. Este algoritmo al igual que el Minimax hace uso de la bsqueda en profundidad asi que para poder encontrar un resultado solo tendremos que considerar los nodos a lo largo de un solo camino.

Entre las principales caractersticas que posee este algoritmo en comparacin con el Minimax tenemos:

- Omisin de expansin de nodos que por sus valores no pueden ser los mejores.
- No afecta al resultado final.
- Permite bsquedas el doble de profundas que el Minimax
- Importancia de orden y no de valores exactos

### Historia Poda Alfa-Beta

Donald Knuth y Ronald Moore (1975) fueron los primeros en analizar a fondo la eficiencia de la poda
alfa-beta, en su libro An analysis of alpha-beta pruning. Es posible calcular en un determinado
juego una decisin correcta sin tener que explorar todos los nodos de un rbol de bsqueda. A
la tcnica que consideramos en particular se le conoce como poda alfa-beta. Aplicada a un rbol
minimax estndar, produce la misma jugada que se obtendra con minimax, pero elimina todas
las ramas que posiblemente no influirn en la decisin final. La eficiencia de alfa-beta depender
del orden como se exploren los sucesores dentro de un rbol, es mejor primero explorar aquellos
sucesores que aparentemente tienen ms posibilidades de ser los mejores.
